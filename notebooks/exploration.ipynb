```python
import pandas as pd
import matplotlib.pyplot as plt

#Cargamos datos limpios
data_path = "../data/processed/cleaned_products.csv"
df = pd.read_csv(data_path)

#Analis y visualizacion
def show_basic_statistics(df):
    print("Estadísticas Básicas:")
    print(df.describe())

#Histograma de precios
plt.hist(df["price"], bins=20, edgecolor="black") 
plt.title("Distribución de precios")
plt.xlabel("Precio") # Añadimos una etiqueta al eje X
plt.ylabel("Frecuencia") # Añadimos una etiqueta al eje y
plt.show() # Mostramos el Histograma.

# Función para mostrar los productos más caros
def show_highest_prices(df):
    print("\nProductos Más Caros:")
    highest_prices = df.nlargest(5, 'price')
    print(highest_prices)

# Llamamos a las funciones definidas anteriormente
show_basic_statistics(df)
show_highest_prices(df)
```
